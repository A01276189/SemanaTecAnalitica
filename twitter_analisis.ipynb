```python
import pandas as pd
data = pd.read_csv('arteDeLaAnalitica/twitter_dataset.csv')
```

```python
data.plot(y='Tweet_ID', kind='box', figsize=(5,3))
data.plot(y='Retweets', kind='box', figsize=(5,3))
data.plot(y='Likes', kind='box', figsize=(5,3))
```

```python
data.plot(y='Tweet_ID', kind='hist', figsize=(5,3))
data.plot(y='Retweets', kind='hist', figsize=(5,3))
data.plot(y='Likes', kind='hist', figsize=(5,3))
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
corr = data[['Tweet_ID', 'Likes', 'Retweets']].corr()
sb.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
