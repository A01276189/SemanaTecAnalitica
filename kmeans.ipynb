```python
import pandas as pd
data = pd.read_csv('arteDeLaAnalitica/twitter_dataset.csv')
```

```python
data.plot.scatter(x='Likes', y='Retweets')
```

```python
from sklearn.cluster import KMeans

X = data[['Likes', 'Retweets']]

kmeans = KMeans(2)

kmeans.fit(X)

centros = kmeans.cluster_centers_
centros
```
